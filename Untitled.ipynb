```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## membaca datasheet

```python
df = pd.read_csv('botak.csv')
```

```python
df
```

```python
# 1. Diagram Histogram Distribusi Usia Responden
plt.figure(figsize=(8, 5))
plt.hist(df['umur'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Usia')
plt.ylabel('Jumlah Responden')
plt.title('Distribusi Usia Responden')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Scatter Plot Hubungan antara Gaji dan Probabilitas Botak
plt.figure(figsize=(8, 5))
sns.scatterplot(x='gaji', y='botak_prob', data=df, alpha=0.5, color='purple')
plt.xlabel('Gaji')
plt.ylabel('Probabilitas Botak')
plt.title('Hubungan antara Gaji dan Probabilitas Botak')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 5. Scatter Plot Hubungan antara Tinggi dan Berat
plt.figure(figsize=(8, 5))
sns.scatterplot(x='tinggi', y='berat', data=df, alpha=0.5, color='orange')
plt.xlabel('Tinggi (cm)')
plt.ylabel('Berat (kg)')
plt.title('Hubungan antara Tinggi dan Berat')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
